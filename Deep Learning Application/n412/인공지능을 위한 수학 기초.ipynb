## 미분법

(문과생의 경우 배우지 않았을 수 있는 내용)

수학에서 미분(영어: derivative, 微分) 또는 도함수(導函數)는 어떤 함수의 정의역 속 각 점에서 함숫값의 변화량과 독립 변숫값의 변화량 비의 극한 혹은 극한들로 치역이 구성되는 새로운 함수

미분법의 공식
- $y = c \space (단, c는 상수) \rightarrow y' = 0$
- $y = x^n \space (단, n는 실수) \rightarrow y' = nx^{n-1}$
- $y = cf(x) \space (단, c는 상수) \rightarrow y' = cf'(x)$
- $y = f(x) \pm  g(x) \space \rightarrow y' = f'(x) \pm g'(x)$
- $y = f(x)g(x) \space \rightarrow y' = f'(x)g(x) + f(x)g'(x)$

자연상수($e$) 지수함수의 미분
- $y = e^x \rightarrow y' = e^x$
- $y = e^{f(x)} \rightarrow y' = f(x) e^{f(x)} $

## 푸리에변환



다항함수, $y=ax+b$의 형태거나, $y=ax^2 + bx + c$ 등의 제곱식을 포함할 수 있는 함수들을 다항함수라고 한다. 초월함수라고 하면, 로그함수, 지수함수, 삼각함수와 같이 다항식으로는 표현할 수 없는 함수를 말한다. 

### 테일러 급수
푸리에 급수, 푸리에 변환을 배우기 위해서는 먼저 테일러 급수, 테일러 변환을 배워야 한다. 위에서 언급한 초월함수를 다항함수 형태로 근사할 수 있지 않을까? 에서 출발된 개념으로 이해하면 된다. 

$f(x) = f(a) + (x-a)f'(a) + {(x-a)^2 \over 2!} f''(a) +  ... $ <br>
$= \Sigma {(x-a)^n \over n!} f^{(n)}(a)$

다음과 같은 형태로 근사한 값으로 변환을 하게 되는 것이다. 
직관적으로 이해해보면, 작은 구간에 대하여 다항식으로 바꾸어 계산을 조금 더 쉽게 해보고자 함이다. 어떤 지점(x=a)에서 초월함수가 근방의 좁은 구간에서는 어떤 다항함수 $g(x)$와 같다고 가정을 해보면, 

$g(x) = w_1 + w_2x + w_3x^2 + ...$ 으로 표현이 가능하고, 우리가 w_n 값들 즉 다항함수의 계수들을 알 수만 있다면, 근사시킬 수 있는 것이다. 조금 더 직관적으로 보면, 선택한 지점(x=a)에서는 함수 값을 그대로 사용하니 맞을 것이고, 그 주변만 맞추면 되는데, 그 주변은 다항함수들을 이용해서 비슷하게 맞추어 준다고 생각하면 되는 것입니다. 분모의 n!을 알아야 하는데 역시 직관적으로 보면, n차 식의 경우 미치는 영향을 조절하기 위함이라고 생각하면 된다. 유도과정은 앞으로의 공부에서 필요하지 않으므로 생략하도록 한다.



### 푸리에 급수 (선택)


자 그러면, 이제 푸리에로 돌아와보자.
다항식의 기준을 삼각함수로 변환을 해보는 것에서 출발한다. 
푸리에 급수에서 근사를 $x^n$의 형태로 했다면, 이번에는 sine, cosine으로 해보자는 것이다. 
\begin{equation}
  g(x) = \Sigma (a_n cos(nw_ot) + b_n sin(nw_ot)) \space (w_0 ={ 2 \pi \over T})
\end{equation}

여기서 복소수 도메인으로 가기위해서는 수식 정리가 필요하다.

- $e^{i \theta} = cos \theta + i sin \theta$
- $e^{-i \theta} = cos \theta - i sin \theta$

두 식을 더하면, 
- $e^{i \theta} + e^{-i \theta} = 2 cos \theta $

반대로 빼주면, 
- $e^{i \theta} - e^{-i \theta} = 2 i sin \theta$

정리해보면, cos, sin은 다음과 같이 표현할 수 있다.
- $ cos \theta = {e^{i \theta} + e^{-i \theta} \over 2}$
- $ sin \theta = {e^{i \theta} - e^{-i \theta} \over 2i}$


\begin{equation}
\begin{split}
  g(x) = \Sigma (a_n {e^{i \theta} + e^{-i \theta} \over 2} + b_n {e^{i \theta} - e^{-i \theta} \over 2i}) \space \space \space \\ = \Sigma ({a_n \over 2}e^{i \theta} + {a_n \over 2}e^{-i \theta}) + \Sigma({b_n \over 2i}e^{i \theta} - {b_n \over 2i}e^{-i \theta}) \\ = \Sigma ({a_n \over 2} + {b_n \over 2i}) e^{i \theta} + \Sigma ({a_n \over 2} - {b_n \over 2i}) e^{-i \theta} \\ [since {1 \over i} = {1 \over i}{i \over i} = {i \over -1} ]\\ = \Sigma ( {a_n - ib_n \over 2})e^{i \theta} + \Sigma ( {a_n + ib_n \over 2})e^{-i \theta}
\end{split}
\end{equation}

- $\theta$ 를 원래 식인 $nw_ot$ 로 바꿔주면, 

\begin{equation}
g(x) = \Sigma ( {a_n - ib_n \over 2})e^{i nw_ot} + \Sigma ( {a_n + ib_n \over 2})e^{-i nw_ot}
\end{equation}

- 주기에 따라 실수부인 $a_n$, 허수부인 $b_n$의 형태로 나눠지는 것을 볼 수 있다.

쉽게 설명하기위해서 `소리`를 예를 들면, 50Hz, 100Hz, 500Hz로 3개의 주파수로 소리를 근사한다고 했을 때, 각각의 주파수마다 계수가 2개(a, b)씩 존재하는 방식으로 바꿀 수 있다는 것이다. 

$(소리) = [a_{50} \space b_{50}] (50Hz) + [a_{100} \space b_{100}] (100Hz) + [a_{500} \space b_{500}] (500Hz)$

의 형태로 근사시켜주는 값이라고 이해하고 넘어가도록 하겠다.

그래서 푸리에변환을 한다는 것은 시간대역을 주파수대역으로 변환한다고 말할 수 있는 것이다.
